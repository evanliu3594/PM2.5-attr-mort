## 5.1 数据导入与清洗

```python
# 调包
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = 'Arial Unicode MS'

# 导入数据
job = pd.read_csv('week5_lagouwang.csv')
# 查看数据属性，检查空值
print('数据属性信息：')
print(job.info(),end='\n\n')

# 将无业务意义的字段"web-scraper-order", "web-scraper-start-url"删除
job.drop(['web-scraper-order','web-scraper-start-url'], axis = 1, inplace = True)

# 查看前五行
job.head()
```

##### 小结
根据探索，确定后续需要对数据进行的，以得到所需目标字段：
1. 拆分position_describe字段，得到单独的岗位描述，并命名为pos_des_1, pos_des_2......
2. 拆分company_describe字段，并分别命名为company_describe、company_business（该字段需要进一步拆分）、staff_number
3. 拆分company_business字段，并分别命名为company_business_1、company_business_2
4. 拆分salary字段，分别命名为为salary_min和salary_max，并增加平均工资字段（取最高工资与最低工资的平均）
5. 拆分requirement字段，并分别命名为require_experience_year, require_education
6. 拆分welfare字段，并将其命名为welfare_1, welfare_2, welfare_3......

## 5.2 数据预处理

```python
##### 拆分position_describe字段，得到单独的岗位描述，并命名为pos_des_1, pos_des_2......#####################
job = pd.concat([job, job['position_describe'].str.split('\n', expand = True)], axis = 1)
# 根据观察一共可以拆分为6组，只保留第一个主要的岗位描述
job = job.drop(['position_describe', 1, 2, 3, 4, 5], axis = 1)
# 更改字段名字
job = job.rename(columns = {0:'job_describe'})

##### 拆分company_describe字段，并分别命名为company_describe、company_business、staff_number###########
job = pd.concat([job, job['company_describe'].str.split('/', expand = True)], axis = 1)
# 重命名
job = job.rename(columns = {0:'company_business', 1:'financing', 2:'staff_number'})
# 删掉被拆分的列
job.drop(['company_describe'], axis = 1, inplace = True)

##### 进一步拆分company_business字段，只保留第一个主要的公司所属行业描述#####################################
job = pd.concat([job, job['company_business'].str.split(',', expand = True)], axis = 1)
# 删除被拆分的列，并只保留拆分后的第一列
job = job.drop(['company_business', 1], axis = 1) 
# 改名
job = job.rename(columns = {0:'company_business'})
# 删除描述中多于的空格
job['company_business'] = job['company_business'].str.replace(' ','')

##### 拆分salary字段，并分别命名为为salary_min和salary_max############################################
# 拆分字段，并删除salary
job = pd.concat([job, job['salary'].str.split('-', expand = True)], axis = 1)
job = job.rename(columns = {0:"salary_min", 1:"salary_max"})
job = job.drop(['salary'], axis = 1)
# 把salary_min和salary_max中的k改为000，并将字段转为int
job['salary_min'] = job['salary_min'].str.replace('k', '000')
job['salary_max'] = job['salary_max'].str.replace('k','000')
job[['salary_min','salary_max']] = job[['salary_min','salary_max']].astype('int')
# 增加平均工资
job['salary_mean'] = (job['salary_min']+job['salary_max'])/2

##### 拆分requirement字段，并分别命名为require_experience_year, require_education##########################
# 字段中有两种分隔符，先用“/”拆分
job = pd.concat([job, job['requirement'].str.split("/", expand = True)], axis = 1)
job = job.rename(columns = {0:"a", 1:"education"})
# 再拆分“\n”
job = pd.concat([job, job['a'].str.split('\n', expand = True)], axis = 1)
job = job.rename(columns = {1:"experience"})
# 删除多于内容
job['experience'] = job['experience'].str.replace(' ','').str.replace('经验','').str.replace('应届毕业生','应届')
# 去掉重复的列
job = job.drop(['requirement', 'a', 0], axis = 1)

##### 拆分welfare字段，并只保留第一个主要的welfare......#####################################
# 将所有“、”“,”“；”“ ”改为“，”，并将前后空格去掉，便于之后拆分
job['welfare'] = job['welfare'].str.replace('、',',').str.replace('，',',').str.replace(' ',',').str.replace('\n',',').str.replace('；',',').str.replace('“','').str.replace('”','')
job = pd.concat([job, job['welfare'].str.split(',', expand = True)], axis = 1)
# 删除被拆分的welfare列，并且只保留拆分后的第一列
job = job.drop(['welfare',1,2,3,4,5], axis = 1)
# 改名字
job = job.rename(columns = {0:'welfare'})

# 查看处理后的结果
job.head()
```
